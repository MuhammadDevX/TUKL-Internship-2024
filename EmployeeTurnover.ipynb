```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

df = pd.read_csv("/content/HR_DataSet.csv")
```

```python
variables = ["Department","salary"]
```

```python
df = pd.get_dummies(df, columns=variables, drop_first=True)
```

```python
X_train = df.drop("left",axis =1)
y_train = df["left"]
X_train
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
model = Sequential(
    [
        Dense(units = 21,activation = "relu"),
        Dense(units = 14,activation ="relu"),
        Dense(units = 7,activation = "relu"),
        Dense(units = 1,activation = "linear")
        ]
)

model.compile(
    optimizer = Adam(learning_rate = 0.01),
    loss = tf.keras.losses.MeanSquaredError()
)
```

```python
X_train , X_test ,y_train,y_test = train_test_split(X_train,y_train,train_size=0.7,random_state=55)
```

```python
model.fit(X_train,y_train,epochs = 25)
```

```python

y_predicted =model.predict(X_test)
```
