```python
train_df = pd.get_dummies(df, columns=['Gender', 'Platform'], drop_first=True)
test_df = pd.get_dummies(test_df, columns=['Gender', 'Platform'], drop_first=True)



# train_df = train_df.dropna()
# test_df = test_df.dropna()


# X_train = np.array(X_train)
# y_train = np.array(y_train)
# for i in range(len(X_train)):
#   for j in range(2):
#     if type(X_train[i][j]) == str:
#       X_train[i][j] = float(X_train[i][j])
# # y_train[0][0]
```

```python
missing_cols_test = set(train_df.columns) - set(test_df.columns)
for col in missing_cols_test:
    test_df[col] = 0
test_df = test_df[train_df.columns]



label_encoder = LabelEncoder()
# len(output_columns)
```

```python
# all_labels = list(train_df['Dominant_Emotion'])
all_labels = list(train_df['Dominant_Emotion']) + list(test_df['Dominant_Emotion'] )
label_encoder.fit(all_labels)
# for i in range(len(X_train[0])):
  # print(type(X_train[0][i]))
```

```python
# y_train_encoded = label_encoder.transform(train_df['Dominant_Emotion'])
y_train_encoded = label_encoder.transform(train_df['Dominant_Emotion'])
y_test_encoded = label_encoder.transform(test_df['Dominant_Emotion'])
model = Sequential([
    Dense(units = 25, activation = "relu"),
    Dense(units = 25, activation = "relu"),
    Dense(units = 25, activation = "relu"),
    Dense(units = 6, activation = "softmax")
    ])
model.compile(
    optimizer = Adam(learning_rate = 0.01),
    loss = "categorical_crossentropy",
)

X_train = train_df.drop('Dominant_Emotion', axis=1)
X_test = test_df.drop('Dominant_Emotion', axis=1)

# Select only numeric columns for scaling
numeric_columns = X_train.select_dtypes(include=[np.number]).columns.tolist()

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[numeric_columns])
X_test_scaled = scaler.transform(X_test[numeric_columns])

# y_train_encoded
# X_train = train_df.drop('Dominant_Emotion', axis=1)
```
